```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import torch as th

path = "/zhome/e3/3/139772/Desktop/ADLCV/Project/ADLCV-AnomalyDetection/results/plots/295_DDIM/"
path_2 = "/zhome/e3/3/139772/Desktop/ADLCV/Project/ADLCV-AnomalyDetection/results/plots/295"

# List of image paths
image_paths = [f'{path}input 0.png', f'{path}input 1.png', f'{path}input 2.png', f'{path}input 3.png', f'{path}ground truth.png',
               f'{path}sampled output 0.png', f'{path}sampled output 1.png' , f'{path}sampled output 2.png', f'{path}sampled output 3.png', f'{path}difftot.png']

fig, axs = plt.subplots(2, 5, figsize=(20, 8))

# Column titles
col_titles = ['T1', 'T1ce', 'T2', 'FLAIR', 'Anomaly Map']

# Row titles
row_titles = ['Input', 'DDIM', 'DDPM']

for i, ax in enumerate(axs.flat):
    img = mpimg.imread(image_paths[i])
    img = np.rot90(img, -1) 
    ax.imshow(img)
    ax.axis('off')  # to hide the axes

    # Set column titles with larger font size
    if i < len(col_titles):
        ax.set_title(col_titles[i], fontsize=20)

# Set row titles with larger font size and closer to the images
for i, ax in enumerate(axs):
    ax[0].text(-0.1, 0.5, row_titles[i], va='center', rotation='vertical', transform=ax[0].transAxes, fontsize=20)

plt.tight_layout()
plt.savefig
plt.show()
```
